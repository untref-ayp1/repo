# El cifrado del César

Tu tarea es [encriptar](https://es.wikipedia.org/wiki/Cifrado_(criptograf%C3%ADa)) un mensaje utilizando el método del "cifrado del César". Este método consiste en desplazar cada letra del mensaje una cantidad fija de posiciones en el alfabeto.

## Reglas
- Cada letra se desplaza hacia adelante en el alfabeto según el número de posiciones especificado.
- Si el desplazamiento supera la letra `'z'`, el conteo continúa desde la letra `'a'`.
- El método no es sensible a mayúsculas y minúsculas: `'A'` se considera equivalente a `'a'`, `'B'` a `'b'`, y así sucesivamente. La salida siempre será en minúsculas.
- Los espacios y otros caracteres no alfabéticos permanecen inalterados en la encriptación.

#### Desplazamiento de 1 posición
- Mensaje original: `zapato`
- Mensaje encriptado: `abqbpv`

### Notas adicionales
- Si el alfabeto se termina antes de completar el desplazamiento necesario, se continúa desde el principio. Por ejemplo, con un desplazamiento de `2` posiciones, la letra `'z'` se convierte en `'b'`.
- Asegurate de que cualquier espacio o puntuación en el mensaje original permanezca inalterado en el mensaje encriptado.

## Concepto matemático: el módulo

> **Recordemos...**  
> El operador módulo (`%`) devuelve el resto de una división entera.  
> Por ejemplo:  
> `7 % 3 = 1`  → porque `7 = 3 * 2 + 1`  
> `10 % 4 = 2`  → porque `10 = 4 * 2 + 2`  
> 👉 Es decir, `a % b` es el resto de dividir `a` por `b`.
