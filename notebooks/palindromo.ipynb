### El Desafío de la Limpieza (Normalización)
Para procesar frases como "La ruta nos aportó otro paso natural", necesitamos que Java ignore:

1. Mayúsculas: `.toLowerCase()`

1. Espacios y Signos.

1. Acentos.

Vamos a crear un método que *limpie* el texto antes de verificar si es palíndromo.

#### Veamos primero como dejar toda la frase en minúsculas (o en mayúsculas, si así lo prefieren)

Investigar: 

- [`String#toLowerCase():String`](https://docs.oracle.com/en/java/javase/25/docs/api/java.base/java/lang/String.html#toLowerCase())
- [`String#toUpperCase():String`](https://docs.oracle.com/en/java/javase/25/docs/api/java.base/java/lang/String.html#toUpperCase())

```java
String conMayusculas = "La Ruta Nos Aportó Otro Paso Natural";
String todoEnMinusculas = conMayusculas.toLowerCase();
String todoEnMayusculas = conMayusculas.toUpperCase();
System.out.println(todoEnMinusculas);
System.out.println(todoEnMayusculas);
System.out.println(conMayusculas);
```

#### Ahora que tenemos toda la frase en minúsculas, quitemos los espacios y signos.

Investigamos el uso del método `replace()` de la clase `String`: [`String#replace(CharSequence, CharSequence):String`](https://docs.oracle.com/en/java/javase/25/docs/api/java.base/java/lang/String.html#replace(java.lang.CharSequence,java.lang.CharSequence))

**Investigar:** [`String.valueOf(char):String`](https://docs.oracle.com/en/java/javase/25/docs/api/java.base/java/lang/String.html#valueOf(char))

```java
System.out.println("Algo, otra cosa".replace(",",""));
//Hemos quitado la , (coma) teemplazandola por un string vacío. 
```

```java
String cadena = "¡Hola, mundo! @Java #Código.";
String aQuitar = " ,./?~!¡@#$%^&*()_+[]{}|;:'";
String cadenaNormalizada = cadena;
for(int i=0; i<aQuitar.length(); i++){
    cadenaNormalizada = cadenaNormalizada.replace(String.valueOf(aQuitar.charAt(i)),"");
}
System.out.println(cadenaNormalizada)
```
